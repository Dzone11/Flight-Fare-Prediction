# EDA
## Change obj_type of date -> time stamp
use to_datetime to convert


```python
train_data["Journey_Day"] = pd.to_datetime(train_data.Date_of_Journey, format = "%d/%m/%Y").dt.day
```

```python
train_data["Journey_Month"] = pd.to_datetime(train_data.Date_of_Journey, format = "%d/%m/%Y").dt.month
```

```python
train_data.head()
```

```python
train_data.drop(["Date_of_Journey"], axis = 1, inplace = True) # drop the date of journe column as we have split it into day and month
```

```python
train_data.head()
```

```python
# similar to date of journey we can extract the hours and minutes from the departure time
train_data["Dep_hour"] = pd.to_datetime(train_data["Dep_Time"]).dt.hour
# extract minutes
train_data["Dep_min"] = pd.to_datetime(train_data["Dep_Time"]).dt.minute
# drop the departure time column
train_data.drop(["Dep_Time"], axis = 1, inplace = True)
```

```python
train_data.head()
```

```python
# similar to date of journey we can extract the hours and minutes from the arrival time
train_data["Arrival_hour"] = pd.to_datetime(train_data.Arrival_Time).dt.hour
# extract minutes
train_data["Arrival_min"] = pd.to_datetime(train_data.Arrival_Time).dt.minute
# drop the arrival time column
train_data.drop(["Arrival_Time"], axis = 1, inplace = True)
```
